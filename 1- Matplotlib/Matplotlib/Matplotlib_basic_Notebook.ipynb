## Functional Method and Object Oriented Methods
**FM <--->OOM**
* FM ile programlama yaparken işlevlerin sonuçlarına güvenilirlik ve verilerin değişmezliği önemlidir.FM mevcut kütüphanedeki parametrelerin her biri bir fonksiyon dur, varolanları kullanır.

* OOM ile programlama yaparken ise nesnelerin özellikleri ve davranışları bir araya getirilerek programın daha düzenli hale gelmesi sağlanır.OOM ise değişkene atayıp öyle devam ediyoruz. 

* FM de plt.subplot(),  OOm da ise  plt.subplots() kullanilir. Burda _s takisina dikkat! 

* FM plt.xlabel(), OOM da is ax.set_xlabel() seklinde kullanilir.Burda ax.set_.. e dikkat! 

## Functional Method 

```python
age = np.arange(25,44, 2.5)
salary = np.arange(40000, 76000, 5000)
salary_2 = [45372, 48876, 53850, 57287, 
            65998, 71496, 75370, 83640]
plt.style.use("seaborn-darkgrid") # plot un arka fonunu degistirdik
plt.figure(figsize = (6, 3))
plt.plot(age, salary)
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("SALARY BY AGE")

plt.grid(color="purple") ## style i gridli olanlarda gridleri modifiye edebiliriz(grid=izgara) mesela: renklerin degisikligi..
                           
plt.show()
```

## OOM(Object Oriented Method)

```python
#FM:
plt.style.use("seaborn-darkgrid")

plt.plot(age, salary)

plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("SALARY BY AGE")

#bu daha mantikli gorunuyor bu case'de, daha az kodla yazmis oluyorum.
```

```python
# OOM: 
fig, ax = plt.subplots()
ax.plot(age, salary, "r")
ax.set_xlabel("Age")
ax.set_ylabel("Salary")
ax.set_title("SALARY BY AGE")

plt.show()

# multiple axes cizecegimiz zaman OOM nin avantaji olabiliyor karmasikligi onluyor.
```

## Functional Method Multi Plot (2 satır 1 sütun toplam 2 akses)

## OOM Multi Plot (2 satır 1 sütun toplam 2 akses)

```python
fig, ax = plt.subplots(2, 1)

ax[0].plot(age, salary, "r")
ax[1].plot(age, salary2, "g")

plt.tight_layout(); # Grafiklerin birbiri içine geçmesini engelliyor.
```

## Functional Method Multi Plot (2 satır 3 sütun toplam 6 akses)

```python
# 2 satir 3 sutunluk 6 grafigi FM ile cizdirme: 

plt.subplot(2, 3, 1)
plt.plot(age, salary, "r")

plt.subplot(2, 3, 2)
plt.plot(age, salary2, "g")

plt.subplot(2, 3, 3)
plt.plot(age, salary, "y")

plt.subplot(2, 3, 4)
plt.plot(age, salary, "bo")

plt.subplot(2, 3, 5)
plt.plot(age, salary, "b")

plt.subplot(2, 3, 6)
plt.plot(age, salary2, "c")

plt.tight_layout()
plt.show()
```

```python
# bir axes in icine birden fazla plot yapmak istedigimizde: eger 4 farkli class varsa bu method kullanilabilirnir.  
x = np.arange(0, 30, 0.2)

fig, ax = plt.subplots(figsize = (8, 4))

ax.plot(x, x-1, color = "green", linewidth = 3, linestyle = "-")
ax.plot(x, x-2, color = "red", lw = 3, linestyle = "-.")
ax.plot(x, x-3, color = "yellow", lw = 3, ls = ":")
ax.plot(x, x-4, color = "purple", linewidth = 3, linestyle = "--")

plt.show()
```

```python
x = np.arange(10)

fig, ax = plt.subplots(2, 2, figsize = (12,6))

plt.suptitle("LineStyles & Markers & Markersize & Color", # OOM da plt.suptitle , FM de ise plt.title() kullanilir!
             fontsize = 20,
             y = 0.95)                                    # y=0.95 basligin posisyonunu ayarlar. bir sey yazmazsam orjinal yerini korur

ax[0, 0].plot(x, "r", marker= "+", markersize = 5)
ax[0, 1].plot(x-2, "y", marker= "o", ms = 5)

ax[1, 0].plot(x-3, "black", marker= "s", ms = 5, lw = 3)
ax[1, 1].plot(x-4, "g", marker= "1", ms = 5)

plt.show()
```

```python
# OOM ile 3 satir 3 sutunda toplam 9 axes cizimi: 

x = np.arange(-50, 50)
y1 = np.power(x, 2)
y2 = np.power(x, 3)
y3 = np.sin(x)
y4 = np.cos(x)
y5 = np.tan(x)
y6 = np.tanh(x)
y7 = np.sinh(x)
y8 = np.cosh(x)
y9 = np.exp(x)



fig1, ax1 = plt.subplots(nrows = 3, ncols = 3, figsize= (20, 20))

ax1[0, 0].plot(x, y1, "tab:blue") # set the color of line chart
ax1[0, 0].set_title("Square Function") # setting title of subplot
ax1[0, 0].set_xlabel("X", fontsize = 18)
ax1[0, 0].set_ylabel("Y", fontsize = 18)

ax1[0, 1].plot(x, y2, "tab:orange") # set the color of line chart
ax1[0, 1].set_title("Cubic Function") # setting title of subplot
ax1[0, 1].set_xlabel("X", fontsize = 18)
ax1[0, 1].set_ylabel("Y", fontsize = 18)

ax1[0, 2].plot(x, y3, "tab:green") # set the color of line chart
ax1[0, 2].set_title("Sine Function") # setting title of subplot
ax1[0, 2].set_xlabel("X", fontsize = 18)
ax1[0, 2].set_ylabel("Y", fontsize = 18)

ax1[1, 0].plot(x, y4, "r-") # set the color of line chart
ax1[1, 0].set_title("Cosine Function") # setting title of subplot
ax1[1, 0].set_xlabel("X", fontsize = 18)
ax1[1, 0].set_ylabel("Y", fontsize = 18)

ax1[1, 1].plot(x, y5, "b-") # set the color of line chart
ax1[1, 1].set_title("Tangent Function") # setting title of subplot
ax1[1, 1].set_xlabel("X", fontsize = 18)
ax1[1, 1].set_ylabel("Y", fontsize = 18)

ax1[1, 2].plot(x, y6, "g-") # set the color of line chart
ax1[1, 2].set_title("Hyperbolic Tangent Function") # setting title of subplot
ax1[1, 2].set_xlabel("X", fontsize = 18)
ax1[1, 2].set_ylabel("Y", fontsize = 18)

ax1[2, 0].plot(x, y7, "m-") # set the color of line chart
ax1[2, 0].set_title("Hyperbolic Sine Function") # setting title of subplot
ax1[2, 0].set_xlabel("X", fontsize = 18)
ax1[2, 0].set_ylabel("Y", fontsize = 18)

ax1[2, 1].plot(x, y8, "y*") # set the color of line chart
ax1[2, 1].set_title("Hyperbolic Cosine Function") # setting title of subplot
ax1[2, 1].set_xlabel("X", fontsize = 18)
ax1[2, 1].set_ylabel("Y", fontsize = 18)

ax1[2, 2].plot(x, y9, "kv") # set the color of line chart
ax1[2, 2].set_title("Exp Function") # setting title of subplot
ax1[2, 2].set_xlabel("X", fontsize = 18)
ax1[2, 2].set_ylabel("Y", fontsize = 18)

plt.show()
```

# Hist Plot
Nümerik değerlerinizin dağılımını yorumlamanızı sağlayan bir grafiktir, histogram çizer.

__Parametreler__

`bins`: Grup sayısını belirler.

`range`: Bin'lerin alt ve üst limitini belirler.

`density`: True olursa grafik yoğunluk belirtir, false olursa frekans belirtir.

`color`: Bin rengi

`alpha`: Saydamlık opaklık

```python
plt.hist(df["Satış"], bins = 30, alpha = 0.5, color = "blue", label = "Satışların Frekansı")
plt.legend()
plt.show()
```

```python
# numpy ile random veri oluşturup plt.hist() ile frekansını görselleştirelim

plt.hist(np.random.randn(1000), bins = 30,
         color = "purple", label = "Frekans", width = 0.15)
plt.legend()
plt.show()
```

```python
plt.hist(np.random.randn(1000), bins = 30,
        alpha = 0.5, color = "purple", label = "Frekans", width = 0.15, range = (-5,5))    # range ile x ekseni dağılımı ayarlanır
plt.legend()
plt.show()
```

```python
plt.hist(np.random.randn(1000), bins = 30,
         color = "magenta", label = "Frekans", width = 0.15, density = True)   # density = True değerlerin yoğunluğunu gösterir 
plt.legend()                                                                   # yukarıdakinde değerlerin sayısını gösteriyordu
plt.show()
```

# Pie Chart

```python
x = np.array([100,100,100,100])
mylabels = ["Apples","Bananas","Cherries","Pear"]
plt.figure(figsize=(6,6))
plt.pie(x,
        labels=mylabels,     # plt.pie icine mylabels vererek uzun uzun yazmaktan kurtuldum
        labeldistance=0.5,   # labeldistance=0.5 label larin merkeze olan uzakligini ayarliyor
        autopct="%.1f")      # autopct="%.1f" kesir kisminda kac basamak gormek istedigimizi ayarlar. 
plt.show()
# labeldistance icin eksi sayi verirsek, tam karsisindaki bolume geciyor,
# yani diagonalinde bulunan degerle yerini degistirir
```

```python
z = np.array([16, 21, 45, 19])
mylabels = ["Apples", "Bananas", "Cherries", "Pear"]
myexplode = [0.1, 0, 0, 0]        # pie chart dan ayiracagimiz dilimin position ini myexplode variableina atadik
# len(mylabels) = len(myexplode)   Elenmanlar index degerlerine gore eslesir.
# Orn: Apple --> 0.1 le eslesiyor. sifirdan farkli oldugu icin disariya cikiyor.  
plt.figure(figsize=(7, 7))
plt.pie(z, labels = mylabels, explode = myexplode, labeldistance=0.75, autopct="%.1f", startangle=45)
plt.show()
# explode = parametresi pie chartdan dilim ayirmamizi saglar, karsinina position yazilir
# startangle = 45 parametresi daire dilimimizin saat yonunun tersinde baslangic noktasindan itibaren belirtilen aci kadar dondurur.
# Baslangic noktasi yatay eksen positive taraftir
```

```python
plt.figure(figsize=(10,8))
color_list = ["gold","yellowgreen","lightcoral","lightskyblue","lightgreen"] # her birini renklendirmek icin
explode_list = [0.1,0.4,0.3,0.2] # dilimlerin ayrilma positionu 
mylabels = ["Apples","Bananas","Cherries","Pear"]
y =np.array([25,35,25,15])
plt.pie(y, 
        labels=mylabels,             # etiket isimleri
        colors=color_list,           # dilim renkleri
        explode=explode_list,        # dilimlerin positionu
        labeldistance=0.75,          # labellarin merkeze olan uzakligi
        autopct="%.1f",              # ondalik kismin basamak sayisi ayarlanir
        startangle=90,               # baslangic noktasindan saat yonunun tersine donme acisi.. 
        pctdistance=0.6,             # label ile value arasindaki uzaklik. 
        shadow=True)                 # golge verip 3 boyutluymus gibi gostermek icin. 
plt.legend()
plt.show()
```

# iris data

```python
df2 = sns.load_dataset("iris")
df2.head()
```

## sepal_length - sepal_width - species / scatter

```python
df2.species.value_counts()
```

```python
# 1. yol
plt.scatter(df2['sepal_length'][df2['species']=="setosa"],df2['sepal_width'][df2['species']=="setosa"])
plt.scatter(df2['sepal_length'][df2['species']=="versicolor"],df2['sepal_width'][df2['species']=="versicolor"])
plt.scatter(df2['sepal_length'][df2['species']=="virginica"],df2['sepal_width'][df2['species']=="virginica"])
plt.legend(["setosa","versicolor","virginica"])
plt.show()
```

```python
# 2. yol
species = df2['species'].unique()       

colors = ['blue', 'green', 'red']        # her bir çiçek türü için farklı bir renk 

for i in range(len(species)):
    plt.scatter(x=df2[df2['species']==species[i]]['sepal_length'], 
                y=df2[df2['species']==species[i]]['sepal_width'],
                c=colors[i], label=species[i])
    
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs. Sepal Width')
plt.legend()
plt.show()
```
