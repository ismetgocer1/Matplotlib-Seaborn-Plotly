## 2: Plot 12 most dangerous cities

```python
kill.head()
```

```python
kill.city.value_counts().head(12)
```

```python
pd.value_counts(kill['city'])[:12]
```

```python
cty = kill.value_counts("city").nlargest(12)
cty
```

```python
yvalue = cty.values
yvalue
```

```python
xindex = cty.index
xindex
```

```python
# cty.values.tolist()
```

```python
fig, ax = plt.subplots(figsize=(10,6))

sns.barplot(x=xindex, y=yvalue)
ax.set_xticklabels(cty.index, color="b", fontsize=8, rotation= 45)
ax.set_title("Most Dangerous City", color="r");
```

```python
# with seaborn barplot
```

```python
fig, ax = plt.subplots(figsize=(8,5))

sns.countplot(x="city", data=kill, order=xindex)
ax.set_xticklabels(cty.index, color="b", fontsize=8, rotation= 45)
ax.set_title("Most Dangerous City", color="r");
```

```python
# with seaborn countplot
```

## 3: Plot 20 most dangerous states

```python
kill_state = kill.state.value_counts().head(20)
kill_state
```

```python
x_index = kill.state.value_counts().head(20).index
x_index
```

```python
y_value = kill.state.value_counts().head(20).values
y_value
```

```python
fig, ax = plt.subplots(figsize=(8,5))

sns.barplot(x= x_index, y= y_value)

ax.set_xticklabels(kill_state.index, color="b", fontsize=8, rotation= 45)
ax.set_title("Most Dangerous States", color="r");
```

```python
# with seaborn barplot
```

```python
fig, ax = plt.subplots(figsize=(8,5))

sns.countplot(x= "state", data=kill, order=x_index)

ax.set_xticklabels(kill_state.index, color="b", fontsize=8, rotation= 45)
ax.set_title("Most Dangerous States", color="r");
```

```python
# with seaborn countplot
```
